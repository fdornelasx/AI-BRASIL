```python
import pandas as pd
```

```python
data = pd.read_csv("imdb-ptbr/imdb-reviews-pt-br.csv")
```

```python
data['text_pt'][0]
#data['sentiment'][22]

#pos > 49451,49446
```

```python
data = data[['text_pt','sentiment']]
```

```python
data.head()
```

```python
X = data['text_pt']
y = data['sentiment']
X.shape, y.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
import nltk
```

```python
vec = CountVectorizer(min_df = 2,tokenizer = nltk.word_tokenize)
```

```python
tokens_data = vec.fit_transform(data['text_pt'])
```

```python
a = ""
for i in "Eu vi o filme recentemente e realmente gostei. Eu me surpreendi e chorei.".split(" "):
    a += str(vec.vocabulary_.get(i.lower())) +" "  #imprime posição do elemento
```

```python
a
```

tokens_data.shape

```python
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
tfidf = TfidfTransformer()
```

```python
tokens_tfidf = tfidf.fit_transform(tokens_data)
```

```python
list(tokens_tfidf.data)[0:][0:10], list(tokens_tfidf.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = tokens_tfidf
y = data['sentiment']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3, shuffle = True)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.naive_bayes import BernoulliNB, MultinomialNB
```

```python
model = MultinomialNB()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
list(y_pred)
```

```python
y_test.values
```

```python
from sklearn.metrics import accuracy_score
#accuracy_score(y_test, y_pred)
accuracy_score(y_test,y_pred)
```
